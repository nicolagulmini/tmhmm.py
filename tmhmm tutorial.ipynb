```python
import tmhmm
from Bio import SeqIO
```

```python
!rm -r spaan
!git clone git://github.com/nicolagulmini/spaan
proteins = list(SeqIO.parse("spaan/data/original_adh_dataset.fasta", "fasta")) # take arbitrary proteins
```

```python
annotations, posteriors = [], []
for protein in proteins: # to do with a group of proteins
    annotation, posterior = tmhmm.predict(protein.seq)
    annotations.append(annotation)
    posteriors.append(posterior)

print(annotations[0])
print(posteriors[0])
```
